```python
#import library and read dataset
import pandas as pd
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

df = pd.read_csv(r'C:\Users\simpo\Downloads\movies.csv')

# pd.set_option('display.max_rows', None)
```

```python
#look at the info of dataset
df.info()
```

```python
# chage datatypes of score,vote budget gross
df['budget'] = df['budget'].astype('Int64')
df['gross'] = df['gross'].astype('Int64')
```

```python
df['votes'] = df['votes'].astype('Int64')
```

```python
# pd.set_option('display.max_rows', None)
```

```python
# checking null value
df.isnull().sum()
```

```python
df.rating
```

```python
# fill null value of rating column

df['rating'].fillna('Not Rated', inplace=True)
```

```python
# fill null value of released column

# df.released
# df['released'].fillna('Unknow', inplace=True)
df['released'] = df['released'].fillna('Unknown')
```

```python
# fill null value of vote column

# df['votes'].fillna(0,inplace=True )
df['votes'] = df['votes'].fillna(0)
```

```python
# fill null value of score coulumn

# df['score'].fillna('0')
df['score'] = df['score'].fillna(0)
```

```python
# fill null value of writer, star, country, budget, gross, company and runtime

df['writer'] = df['writer'].fillna('Unknow')
df['star'] = df['star'].fillna('Unknow')
df['country'] = df['country'].fillna('Unknow')
df['budget'] = df['budget'].fillna(0)
df['gross'] = df['gross'].fillna(0)
df['company'] = df['company'].fillna('Unknow')
df['runtime'] = df['runtime'].fillna(0)

# df['writer'].fillna('Unknow')
# df['star'].fillna('Unknow')
# df['country'].fillna('Unknow')
# df['budget'].fillna(0)
# df['gross'].fillna(0)
# df['company'].fillna('Unknow')
# df['runtime'].fillna(0)
```

```python
#drop year columns for create exactly year

df = df.drop('year', axis=1)
```

```python
# have some released year was not = year columns
# released  7666 non-null   object
df['year'] = df['released'].str.split(',').str[1].str.strip().str[:4]
df['year'] = df['year'].astype('Int64')
```

```python
#scatter plot to see Budget vs Gross

plt.scatter(x=df['budget'],y=df['gross'],color='hotpink')
plt.title('Budget vs Gross')
plt.xlabel('Budget')
plt.ylabel('Gross')
plt.show
```

```python
#sort dataframe by use gross value desc

df.sort_values(by=['gross'], ascending=False)
```

```python
#regression plot 

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "pink"}, line_kws={"color": "blue"})
```

```python
#see correlation of dataframe and select only number type as int, float

number = df.select_dtypes(include=['number'])
number.corr(method='pearson')
#corr method pearson, kendall, spearman
# df.corr()
```

```python
#create heatmap from of correlation

matrix = number.corr(method='pearson')

sns.heatmap(matrix, annot=True)
plt.title('Correlation Matrix')

plt.xlabel('Movie Industry')
plt.ylabel('Movie Industry')
plt.show()
```

```python
#change the columns that have datatype as a object to category and use cat.codes to generate number

df_toNum = df
for column in df_toNum:
    if(df_toNum[column].dtypes =='object'):
        df_toNum[column] = df_toNum[column].astype('category')
        df_toNum[column] = df_toNum[column].cat.codes
df_toNum.sort_values(by=['gross'], ascending=False)
```
